Primera Parte 

Dado que $G(i,j,\theta)$ es una matriz diagonal con 1's en la diagonal excepto en las columnas $i,j$, se sigue que $p_l= v_lg_l = v_i$ con $l\not= i,j$. Ahora, si $l= i$, entonces $p_l= v_icos(\theta)- v_jsin(\theta)$, de manera similar, si $l= j$ entonces $p_l= v_jcos(\theta)+ v_isin(\theta)$

```python
def product(V, n, i, j, c, s):
  v_i= V[:,i].copy()
  v_j= V[:,j].copy()
  for k in range(n):
    V[k][i]= c*v_i[k]- s*v_j[k]
    V[k][j]= c*v_j[k]+ s*v_i[k]
  return V
```
